```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def tambah(self, data):
        baru = Node(data)
        if not self.head:
            self.head = baru
            return

        last = self.head
        while last.next:
            last = last.next

        last.next = baru
        baru.prev = last

    def sisip_tengah(self, data, setelah):
        baru = Node(data)
        cari = self.head

        while cari and cari.data != setelah:
            cari = cari.next

        if not cari:
            return

        baru.next = cari.next
        baru.prev = cari

        if cari.next:
            cari.next.prev = baru

        cari.next = baru

    def hapus(self, keyword):
        cari = self.head

        while cari and cari.data != keyword:
            cari = cari.next

        if not cari:
            return

        if cari.prev:
            cari.prev.next = cari.next
        else:
            self.head = cari.next

        if cari.next:
            cari.next.prev = cari.prev

    def cetak(self):
        isi = []
        current = self.head

        while current:
            isi.append(str(current.data))
            current = current.next

        print(", ".join(isi))


dll = DoubleLinkedList()

dll.tambah("Januari")
dll.tambah("Februari")
dll.tambah("Maret")
print("List awal:")
dll.cetak()

print("=======================================")
dll.sisip_tengah("April", "Februari")
print("Setelah sisip April setelah Februari:")
dll.cetak()

print("=======================================")
dll.hapus("Februari")
print("Setelah hapus Februari:")
dll.cetak()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def tambah(self, data):
        baru = Node(data)

        if not self.head:
            self.head = self.tail = baru
            baru.next = self.head
            return

        self.tail.next = baru
        baru.next = self.head
        self.tail = baru

    def tambah_akhir(self, data):
        self.tambah(data)

    def hapus_akhir(self):
        if not self.head:
            return

        if self.head == self.tail:
            self.head = self.tail = None
            return

        cari = self.head
        while cari.next != self.tail:
            cari = cari.next

        cari.next = self.head
        self.tail = cari

    def cetak(self):
        if not self.head:
            print("Kosong")
            return

        isi = []
        current = self.head

        while True:
            isi.append(str(current.data))
            current = current.next
            if current == self.head:
                break

        print(", ".join(isi) + " (kembali ke awal)")


cll = CircularLinkedList()

cll.tambah("Januari")
cll.tambah("Februari")
cll.tambah("Maret")
print("List awal:")
cll.cetak()

print("--------------------------------------------")
cll.tambah_akhir("April")
print("Setelah tambah April di akhir:")
cll.cetak()

print("--------------------------------------------")
cll.hapus_akhir()
print("Setelah hapus akhir:")
cll.cetak()
```
