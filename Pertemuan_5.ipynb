```python
import math

def jumpSearch(arr, x, n):
  step = math.sqrt(n)
  prev = 0
  while arr[int(min(step, n)-1)] < x:
    prev = step
    step += math.sqrt(n)
    if prev >= n:
      return -1

  while arr[int(prev)] < x:
    prev += 1
    if prev == min(step, n):
      return -1

  if arr[int(prev)] == x:
    return prev

  return -1

arr = [3,5,7,9,12,13,26,29,34,35,36,38,45,47,49,51,54,76,89,90,93,95,97,100,110]
x = 100
n = len(arr)

index = jumpSearch(arr, x, n)

print("Number", x, "is at index","%.0f"%index)
```

```python
def printTwoElements(arr):
  n = len(arr)
  temp = [0] * n
  repeatingNumber = -1
  missingNumber = -1

  for i in range(n):
    temp[arr[i] - 1] += 1
    if temp[arr[i] - 1] > 1:
      repeatingNumber = arr[i]
  for i in range(n):
    if temp[i] == 0:
      missingNumber = i + 1
      break

  print("The repeating number is", repeatingNumber, ".")
  print("The missing number is", missingNumber, ".")

arr = [9,2,8,4,3,5,6,1,8]
printTwoElements(arr)
```

```python
def cari_pasangan(arr, X):
    n = len(arr)
    ukuran = max(arr) + X + 1
    temp = [0] * ukuran

    for i in range(n):
        cari = X - arr[i]
        if cari >= 0 and temp[cari] == 1:
            return True, (cari, arr[i])
        temp[arr[i]] = 1
    return False, None

arr1 = [11, 15, 6, 8, 9, 10]
X1 = 16
hasil1, pasangan1 = cari_pasangan(arr1, X1)
print(f"Output: {hasil1}")
if hasil1:
    print(f"Terdapat Pasangan {pasangan1} dengan jumlah {X1}")
print()

arr2 = [11, 15, 26, 38, 9, 10]
X2 = 35
hasil2, pasangan2 = cari_pasangan(arr2, X2)
print(f"Output: {hasil2}")
if hasil2:
    print(f"Terdapat Pasangan {pasangan2} dengan jumlah {X2}")
print()

arr3 = [11, 15, 26, 38, 9, 10]
X3 = 45
hasil3, pasangan3 = cari_pasangan(arr3, X3)
print(f"Output: {hasil3}")
if not hasil3:
    print("Tidak ada pasangan dengan jumlah", X3)
```
